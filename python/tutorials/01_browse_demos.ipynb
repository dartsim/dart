# Browse DART Demos

Walk through DART's headless Python demos directly from a Jupyter / Colab
notebook. The notebook **imports** the scene modules from
`python/examples/demos` (no scene-logic duplication) — see PLAN-103.

**Local prerequisites**: a built `dartpy` on `PYTHONPATH`. Inside the DART repo:

```bash
pixi run build
PYTHONPATH=build/default/cpp/Release/python:python jupyter notebook python/tutorials
```

**Colab**: see `python/tutorials/README.md` for the cell pattern that installs the
DART 7 dartpy wheel and fetches the `examples.demos` package from the repo (the
exact Colab plumbing is owned by PLAN-012).

````python
import os, pathlib, sys

# Make `examples.demos` importable when run from this notebook's location.
here = pathlib.Path(os.getcwd()).resolve()
for candidate in (here, here.parent):
    if (candidate / 'examples' / 'demos').is_dir():
        if str(candidate) not in sys.path:
            sys.path.insert(0, str(candidate))
        break

import dartpy  # noqa: F401
from examples.demos import make_demo_scenes, run
````

## List the catalog

````python
scenes = make_demo_scenes()
for scene in scenes:
    print(f'{scene.category:<28s} {scene.id:<24s} {scene.title}')
````

## Build and step a scene

Pick any scene by id and run a few deterministic steps. Each scene exposes a
`build()` factory returning a `SceneSetup` with the `world` and an optional
custom `step(n)` callable.

````python
scenes_by_id = {scene.id: scene for scene in scenes}
hello = scenes_by_id['hello_world']
setup = hello.build()

for _ in range(60):
    setup.world.step()

falling_box = setup.world.get_skeleton('falling_box')
print('falling_box positions after 60 steps:', list(falling_box.get_positions()))
````

## Run the registered scene cycle

The CLI runner is callable from Python too — useful for sanity-checking the
whole catalog from a notebook.

````python
run(['--cycle-scenes', '--frames', '2', '--headless'], scenes)
````
